```python
import numpy as np
import matplotlib.pyplot as plt
#from matplotlib import pyplot as plt
```

```python
def singleton(x, x0):
    if (type(x) is int) or (type(x) is float):
        if x == x0:
            m = 1.0
        else:
            m = 0.0
        return m
    elif (type(x) is np.ndarray):
        m = np.zeros(x.size)
        for i in range(x.size):
            if x[i] == x0:
                m[i] = 1.0
            else:
                m[i] = 0.0
        return m
    else:
        return -1
```

```python
x = np.linspace(0.0,10.0,1001)
x0 = 3.9
m = singleton(x,x0)
xtest = 3.9
#print("Con x0", x0, "el valor singleton de xtest es", singleton(xtest,x0))
print("E. Mora\n" + "Enero 2022\n" + "Variables Lingüísticas en Python")
plt.title("Función de membresía - Singleton")
plt.xlabel("x0")
plt.ylabel("μ (X)")
plt.plot(x,m)
plt.show()
```

```python
def trim(x, param):
    #para = [a,b,c]
    # a <= b <= c
    a = float(param[0])
    b = float(param[1])
    c = float(param[2])
    if (a<= b) and (b <= c):
        if (type(x) is int) or (type(x) is float):
            if x < a:
                m = 0.0
            elif (a <= x) and (x <= b):
                m = (x-a) / (b-a)
            elif (b <= x) and (x <= c):
                m = (c-x) / (c-b)
            else:
                m = 0.0
            return m
        else:
            m = np.zeros(x.size)
            for i in range(x.size):
                if x[i] <= a:
                    m[i] = 0.0
                elif (a <= x[i]) and (x[i] <= b):
                    m[i] = (x[i] - a) / ( b-a )
                elif (b <= x[i]) and (x[i] <= c):
                    m[i] = (c - x[i]) / ( c-b )
                else:
                    m[i] = 0.0
            return m
    else:
        return -1
                    
```

```python
x = np.linspace(-10,10,1001)
a,b,c = -4,-1, 8
xtest = 2.7
m = trim(x, [a,b,c])
print("E. Mora\n" + "Enero 2022\n" + "Variables Lingüísticas en Python")
plt.title("Función de membresía - Triangular")
plt.xlabel("x-test")
plt.ylabel("μ (X)")
plt.plot(x,m)
print("para xtest =", xtest, "m(x) = ", trim(xtest, [a,b,c]
                                            ))
```
